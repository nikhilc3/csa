# Implementing Linked Lists

Before starting this I had no clue what a linked list was, so I had to do some research. 

A linked list is a fundamental data structure used in computer science to organize items sequentially. Unlike arrays, linked lists are not stored in contiguous memory locations; instead, each element (commonly called a "node") contains its own data and a reference (or link) to the next node in the sequence. This structure allows for efficient insertion and deletion of elements.

In Java, linked lists can be implemented using the linkedlist class from the java.util package or by implementing a custom linked list.

In a single Linked list, navigation is only forward, whereas in a double linked list, the navigation is forward and backwards, cicular linked list is that the lasst element is linked to the first element



Single linked list is a list made up of nodes: These nodes are two parts
- The nodes are the Data and the link
- nodes do not need to be in sequential order

```java
// Class to represent each node in the linked list
class Node {
    int data;   // Data element of the node
    Node next;  // Reference to the next node in the list ( the link)

    // Constructor to create a new node with given data
    public Node(int data) {
        this.data = data;
        this.next = null; // Next is initialized to null
    }
}

// Class to represent the entire linked list
class LinkedList {
    Node head;  // Reference to the first node of the list

    // Constructor to create an empty linked list
    public LinkedList() {
        this.head = null; // Initializes the head to null indicating the list is empty
    }

    // Method to add a new node with the specified data at the end of the list
    public void add(int data) {
        Node newNode = new Node(data); // Creates a new node
        if (head == null) {
            head = newNode; // If the list is empty, the new node becomes the head
        } else {
            Node current = head; // Start from the head
            while (current.next != null) { // Traverse to the end of the list
                current = current.next;
            }
            current.next = newNode; // Link the last node to the new node
        }
    }

    // Method to display all nodes in the list from head to end
    public void display() {
        Node current = head; // Start from the head
        while (current != null) {
            System.out.print(current.data + " "); // Print the data of each node
            current = current.next; // Move to the next node
        }
        System.out.println(); // Move to the next line after printing all nodes
    }
}

// Main class to run the linked list operations
public class Main {
    public static void main(String[] args) {
        LinkedList list = new LinkedList(); // Create a new linked list
        list.add(1); // Add elements to the list
        list.add(2);
        list.add(3);
        list.display(); // Display the list
    }
}
// Class to represent each node in the linked list
class Node {
    int data;   // Data element of the node
    Node next;  // Reference to the next node in the list

    // Constructor to create a new node with given data
    public Node(int data) {
        this.data = data;
        this.next = null; // Next is initialized to null
    }
}

// Class to represent the entire linked list
class LinkedList {
    Node head;  // Reference to the first node of the list

    // Constructor to create an empty linked list
    public LinkedList() {
        this.head = null; // Initializes the head to null indicating the list is empty
    }

    // Method to add a new node with the specified data at the end of the list
    public void add(int data) {
        Node newNode = new Node(data); // Creates a new node
        if (head == null) {
            head = newNode; // If the list is empty, the new node becomes the head
        } else {
            Node current = head; // Start from the head
            while (current.next != null) { // Traverse to the end of the list
                current = current.next;
            }
            current.next = newNode; // Link the last node to the new node
        }
    }

    // Method to display all nodes in the list from head to end
    public void display() {
        Node current = head; // Start from the head
        while (current != null) {
            System.out.print(current.data + " "); // Print the data of each node
            current = current.next; // Move to the next node
        }
        System.out.println(); // Move to the next line after printing all nodes
    }
}

// Main class to run the linked list operations
public class Main {
    public static void main(String[] args) {
        LinkedList list = new LinkedList(); // Create a new linked list
        list.add(1); // Add elements to the list
        list.add(2);
        list.add(3);
        list.display(); // Display the list
    }
}
Main.main(null);
```

# Bubble Sort

```java
class Node {
    int data; // Data value of the node
    Node next; // Reference to the next node in the list

    // Constructor to initialize the node with data
    public Node(int data) {
        this.data = data;
        this.next = null;
    }
}

class LinkedList {
    Node head; // Reference to the first node of the linked list

    public LinkedList() {
        this.head = null; // Initializes an empty linked list
    }

    public void insert(int data) {
        // Method to insert a node in sorted order
        Node newNode = new Node(data); // Create a new node with given data
        if (head == null || head.data >= newNode.data) {
            // If the list is empty or the new node should be the new head
            newNode.next = head;
            head = newNode;
        } else {
            // Find the correct position to insert the new node
            Node current = head;
            while (current.next != null && current.next.data < newNode.data) {
                current = current.next;
            }
            newNode.next = current.next;
            current.next = newNode;
        }
    }

    public void display() {
        // Method to display all nodes in the linked list
        Node current = head;
        while (current != null) {
            System.out.print(current.data + " ");
            current = current.next;
        }
        System.out.println();
    }

    public void bubbleSort() {
        // Method to sort the linked list using bubble sort
        boolean wasChanged;
        do {
            Node current = head; // Start from the first node
            Node previous = null; // Previous node starts as null
            Node next = head.next; // Next node is the second node
            wasChanged = false;

            while (next != null) {
                if (current.data > next.data) {
                    // If current node's data is greater than the next node's data, swap them
                    wasChanged = true; // Set flag to true indicating a swap occurred

                    if (previous != null) {
                        // Standard swap when not at the head
                        Node sig = next.next;
                        previous.next = next;
                        next.next = current;
                        current.next = sig;
                    } else {
                        // Swap involving the head
                        Node sig = next.next;
                        head = next;
                        next.next = current;
                        current.next = sig;
                    }

                    previous = next; // Move previous to this node
                    next = current.next; // Move next to the next node
                } else {
                    // Move the pointers forward
                    previous = current;
                    current = next;
                    next = next.next;
                }
            }
        } while (wasChanged); // Continue until no swaps are needed
    }
}

public class Main {
    public static void main(String[] args) {
        LinkedList list = new LinkedList();
        list.insert(5);
        list.insert(3);
        list.insert(8);
        list.insert(1);
        list.insert(9);

        System.out.println("After sorting:");
        list.display(); 
    }
}
Main.main(null);
```

# Selection Sort

Challenges: I was not understanding how to do it and struggled as I was not using a nested for loop and was iterating with just i. Then I used external resources and saw that we needed to use a nested for loop and then I understood the logic of why.
If we try to find the smallest element using only the outer loop (i), we wouldn't be able to properly compare and swap elements because:
- The outer loop itself just iterates through the list.
- We need a separate mechanism (inner loop) to look ahead and find the smallest element in the unsorted part of the list.


```java
class Node {
    int data; // Data value of the node
    Node next; // Reference to the next node in the list

    // Constructor to initialize the node with data
    public Node(int data) {
        this.data = data;
        this.next = null;
    }
}

class LinkedList {
    Node head; // Reference to the first node of the linked list

    public LinkedList() {
        this.head = null; // Initializes an empty linked list
    }

    public void insert(int data) {
        // Method to insert a node at the end
        Node newNode = new Node(data); // Create a new node with given data
        if (head == null) {
            head = newNode;
        } else {
            Node current = head;
            while (current.next != null) {
                current = current.next;
            }
            current.next = newNode;
        }
    }

    public void display() {
        // Method to display all nodes in the linked list
        Node current = head;
        while (current != null) {
            System.out.print(current.data + " ");
            current = current.next;
        }
        System.out.println();
    }

    public void selectionSort() {
        // Method to sort the linked list using selection sort, 
        // This nested for loop, i starts at the minimum and j iterates through evertying after that, then once found the minimum it switches, after that i goes to next node and j iterates through everything after that and finds min and swaps. 
        for (Node i = head; i != null; i = i.next) { // iterate through each node i in the list
            Node min = i;
            for (Node j = i.next; j != null; j = j.next) { // for each node i find the node min with the smallest data value in the remaining list including i
                if (j.data < min.data) {
                    min = j;
                }
            }
            if (min != i) {
                int temp = i.data;
                i.data = min.data;
                min.data = temp;
            }
        }
    }
}

public class Main {
    public static void main(String[] args) {
        LinkedList list = new LinkedList();
        list.insert(5);
        list.insert(3);
        list.insert(8);
        list.insert(1);
        list.insert(9);

        System.out.println("Before sorting:");
        list.display(); 

        list.selectionSort(); 

        System.out.println("After sorting:");
        list.display();
    }
}
Main.main(null);
```

# Capabilities of Object Overrides

To utilize the capabilities of object overrides with `toString()` and `compareTo()` methods for sorting, we can implement the `Comparable` interface in the `Node` class and override the `compareTo()` method to define the natural ordering of nodes based on their data. Additionally, we can override the `toString()` method in the `Node` class to provide a string representation of the node's data, as seen in the revised code below:

```java
class Node implements Comparable<Node> {
    int data; // Data value of the node
    Node next; // Reference to the next node in the list

    // Constructor to initialize the node with data
    public Node(int data) {
        this.data = data;
        this.next = null;
    }

    @Override
    public int compareTo(Node other) {
        // Compare nodes based on their data values
        return Integer.compare(this.data, other.data);
    }

    @Override
    public String toString() {
        // Convert node data to string
        return Integer.toString(data);
    }
}

class LinkedList {
    Node head; // Reference to the first node of the linked list

    public LinkedList() {
        this.head = null; // Initializes an empty linked list
    }

    public void insert(int data) {
        // Method to insert a node in sorted order
        Node newNode = new Node(data); // Create a new node with given data
        if (head == null || head.data >= newNode.data) {
            // If the list is empty or the new node should be the new head
            newNode.next = head;
            head = newNode;
        } else {
            // Find the correct position to insert the new node
            Node current = head;
            while (current.next != null && current.next.data < newNode.data) {
                current = current.next;
            }
            newNode.next = current.next;
            current.next = newNode;
        }
    }

    public void display() {
        // Method to display all nodes in the linked list
        Node current = head;
        while (current != null) {
            System.out.print(current + " ");
            current = current.next;
        }
        System.out.println();
    }

    public void bubbleSort() {
        // Method to sort the linked list using bubble sort
        boolean wasChanged;
        do {
            Node current = head; // Start from the first node
            Node previous = null; // Previous node starts as null
            Node next = head.next; // Next node is the second node
            wasChanged = false;

            while (next != null) {
                if (current.compareTo(next) > 0) {
                    // If current node's data is greater than the next node's data, swap them
                    wasChanged = true; // Set flag to true indicating a swap occurred

                    if (previous != null) {
                        // Standard swap when not at the head
                        Node sig = next.next;
                        previous.next = next;
                        next.next = current;
                        current.next = sig;
                    } else {
                        // Swap involving the head
                        Node sig = next.next;
                        head = next;
                        next.next = current;
                        current.next = sig;
                    }

                    previous = next; // Move previous to this node
                    next = current.next; // Move next to the next node
                } else {
                    // Move the pointers forward
                    previous = current;
                    current = next; 
                    next = next.next;
                }
            }
        } while (wasChanged); // Continue until no swaps are needed
    }
}

public class Main {
    public static void main(String[] args) {
        LinkedList list = new LinkedList();
        list.insert(5);
        list.insert(3);
        list.insert(8);
        list.insert(1);
        list.insert(9);

        System.out.println("Before sorting:");
        list.display(); // Display list before sorting

        list.bubbleSort(); // Sort the list using bubble sort

        System.out.println("After sorting:");
        list.display(); // Display list after sorting
    }
}
Main.main(null);
```

```java
import java.util.function.Function;

class Node<T extends Comparable<T>> {
    T data; // Data value of the node
    Node<T> next; // Reference to the next node in the list

    // Constructor to initialize the node with data
    public Node(T data) {
        this.data = data;
        this.next = null;
    }

    @Override
    public String toString() {
        return data.toString();
    }
}

class LinkedList<T extends Comparable<T>> {
    Node<T> head; // Reference to the first node of the linked list
    Function<Node<T>, Comparable> keyExtractor;

    public LinkedList() {
        this.head = null; // Initializes an empty linked list
        this.keyExtractor = node -> node.data; // Default key extractor
    }

    public void insert(T data) {
        Node<T> newNode = new Node<>(data); // Create a new node with the given data
        if (head == null) {
            head = newNode; // If the list is empty, the new node becomes the head
        } else {
            Node<T> current = head;
            while (current.next != null) { // Traverse to the end of the list
                current = current.next;
            }
            current.next = newNode; // Link the last node to the new node
        }
    }

    public void display() {
        Node<T> current = head; // Start from the head
        while (current != null) {
            System.out.print(current + " "); // Print the data of each node
            current = current.next; // Move to the next node
        }
        System.out.println(); // Move to the next line after printing all nodes
    }

    public void bubbleSort() {
        boolean wasChanged;
        do {
            Node<T> current = head;
            Node<T> previous = null;
            Node<T> next = head.next;
            wasChanged = false;

            while (next != null) {
                if (keyExtractor.apply(current).compareTo(keyExtractor.apply(next)) > 0) {
                    // Swap the nodes
                    wasChanged = true;

                    if (previous != null) {
                        Node<T> sig = next.next;
                        previous.next = next;
                        next.next = current;
                        current.next = sig;
                    } else {
                        Node<T> sig = next.next;
                        head = next;
                        next.next = current;
                        current.next = sig;
                    }

                    previous = next;
                    next = current.next;
                } else {
                    previous = current;
                    current = next;
                    next = next.next;
                }
            }
        } while (wasChanged);
    }

    // Method to change the sort key extractor
    public void changeSortKey(Function<Node<T>, Comparable> keyExtractor) {
        this.keyExtractor = keyExtractor;
    }
}

public class Main {
    public static void main(String[] args) {
        LinkedList<Integer> list = new LinkedList<>();
        list.insert(5);
        list.insert(3);
        list.insert(8);
        list.insert(1);
        list.insert(9);

        System.out.println("Before sorting:");
        list.display(); // Display the list before sorting

        list.bubbleSort(); // Sort the list

        System.out.println("After sorting by data:");
        list.display(); // Display the list after sorting

        // Change sorting key to the next node's data
        list.changeSortKey(node -> node.next != null ? node.next.data : Integer.MAX_VALUE);
        list.bubbleSort();
        System.out.println("After sorting by next node's data:");
        list.display(); // Display the list after sorting
    }
}
Main.main(null);
```

So basically a key extractor will extract a part of the node and use that part to sort the data, in this case we used the data as the key extractor. 

Generic T allows you to write classes, interfaces, etc with placeholder for type of data so integer and string

the key extractor function takes a node and gives us the value we want to use for sorting the nodes. 

By using key extractor we flexibly decide we want to comprae the nodes based on their data, their link, their neighbor or whatever.

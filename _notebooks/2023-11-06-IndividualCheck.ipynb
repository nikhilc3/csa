# Individual Checks

### Individual Code: Body Diagram 
I've created an interactive web page titled "Interactive Body Diagram," which is accessible at the /bodydiagram permalink. The page features a detailed SVG body diagram where users can interact with different parts labeled as "cancer." These parts are draggable and can be moved into a designated area on the screen.

The body diagram is styled to be centered on the page with a white background, and the cancer spots are colored red to stand out. I've used CSS to style the SVG elements, giving them a consistent look and feel with strokes and fills. There's also a modal component that remains hidden until activated, which displays detailed information about the cancer type that the user interacts with.

I've implemented JavaScript to handle the drag-and-drop functionality and to fetch data from an external API. When a user clicks on a cancer spot, the script fetches data from https://kpns.stu.nighthawkcodingsociety.com/api/cancers/byCancerType/ using the cancer type as a parameter. The fetched data includes the number of people affected, death rate, average recovery time, and symptoms, which are then displayed in the modal.

The script also listens for mouse events to handle the dragging of cancer spots, displaying the modal with fetched data, and closing the modal. I've made sure that the draggable elements are converted from SVG circles to absolute-positioned divs to make them easier to manipulate outside the SVG canvas.

In summary, the page is designed to be an educational tool that provides an interactive way to learn about different types of cancer through a visual representation of the human body. Users can drag the red dots representing different cancers into a gray box, and upon clicking these dots, they can get more information about each type of cancer in a modal popup.

# Trimester Reflection on Java and JavaScript Projects

## Java Backend Development Reflection

This trimester, I focused on developing a RESTful API using Java with the Spring Boot framework. The project involved creating a comprehensive backend system for managing cancer data, which included CRUD operations for cancer records.

### Key Learnings:

- **Spring Boot Framework:** I gained a deeper understanding of the Spring Boot framework, particularly in creating RESTful services and managing dependencies.
- **Data Management:** I learned how to interact with a JPA repository to perform database operations, which was crucial for managing the cancer data records.
- **API Design:** I developed skills in designing API endpoints that are intuitive and REST-compliant, ensuring that the frontend could easily interact with the backend.
- **Error Handling:** I implemented proper error handling to return appropriate HTTP status codes, which improved the reliability of the API.

### Challenges:

- **Complex Queries:** Crafting JPA queries to filter data based on various criteria was challenging but rewarding, as it improved the API's flexibility.
- **CORS Configuration:** Understanding and implementing Cross-Origin Resource Sharing (CORS) was initially a hurdle but was essential for the frontend to communicate with the backend. I had a really big cors error and it was extremely hard to fix but I did fix it.

## JavaScript Frontend Development Reflection

On the frontend, I worked with vanilla JavaScript to create an interactive body diagram that allows users to learn about different types of cancers.

### Key Learnings:

- **DOM Manipulation:** I became proficient in manipulating the Document Object Model (DOM) to create dynamic and interactive user experiences.
- **Event Handling:** I learned to handle various events such as clicks, mouse movements, and drags, which was pivotal for the interactive aspect of the body diagram.
- **Asynchronous JavaScript:** I used `fetch` API for asynchronous data retrieval from the backend, which was essential for displaying cancer-related data.

### Challenges:

- **SVG Interaction:** Working with SVG elements to create interactive diagrams was initially complex, especially when making them responsive to user actions.
- **UI/UX Design:** Ensuring that the user interface was intuitive and user-friendly required careful design and testing.

## Overall Reflection

The past trimester has been a journey of growth and learning. Working on both the backend and frontend allowed me to appreciate the full stack of web development. I encountered challenges that pushed me to find creative solutions and deepen my understanding of both Java and JavaScript.

The experience has not only enhanced my technical skills but also my problem-solving abilities and attention to detail. As I move forward, I plan to build on this foundation, exploring more advanced topics and technologies to create even more robust and user-friendly applications.
